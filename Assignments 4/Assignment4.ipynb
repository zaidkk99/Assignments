```python
#  Find and display the largest number of a list without using built-in function max(). Your program 
# should ask the user to input values in list from keyboard. 
```

```python
lst = input("Enter a List: ").split()
largest = lst[0]
for x in lst:
    if x>largest:
        largest = x
print(largest)
```

```python
#  Write a program that reads a string from the user containing a date in the form mm/dd/yyyy.
# It should print the date in the form March 12, 2021. 
```

```python
months = {"01": "January", "02": "February", "03": "March",
        "04": "April", "05": "May", "06": "June",
        "07": "July", "08": "August", "09": "September",
        "10": "October", "11": "November", "12": "December"}
date_str = input("Enter date (mm/dd/yyyy): ")
mm, dd, yyyy = date_str.split("/")
print(f"{months[mm]} {int(dd)}, {yyyy}")
```

```python
'''  Write a program with a function that accepts a string from keyboard and create a new string after 
converting character of each word capitalized. For instance, if the sentence is "stop and smell the roses." 
the output should be "Stop And Smell The Roses" '''
```

```python
def capitalize_words(str):
    words = str.split()
    capitalized_words = [word.capitalize() for word in words]
    return " ".join(capitalized_words)
input = input("Enter a string: ")
result = capitalize_words(input)
print("Converted sentence:", result)
```

```python
''' . Find the sum of each row of matrix of size m x n. For example for the following matrix output will be 
like this : 
Sum of row 1 = 32 
Sum of row 2 = 31 
Sum of row 3 = 63 '''
```

```python
def row_sums(matrix):
    for i, row in enumerate(matrix, start=1):
        print(f"Sum of row {i} = {sum(row)}")
matrix = [
    [5, 7, 10, 10],  
    [8, 4, 9, 10],   
    [20, 10, 15, 18]  
]
row_sums(matrix)
```

```python
# Write a program to add two matrices of size n x m. 
```

```python
def add_matrices(matrix1, matrix2):
    result = []
    for i in range(len(matrix1)):  
        row = []
        for j in range(len(matrix1[0])):
            row.append(matrix1[i][j] + matrix2[i][j])
        result.append(row)
    return result
matrix1 = [
    [5, 7, 10],
    [8, 4, 9],
    [20, 10, 15]
]

matrix2 = [
    [9, 8, 7],
    [6, 5, 4],
    [3, 2, 1]
]
result = add_matrices(matrix1, matrix2)
print("Resultant Matrix:")
for row in result:
    print(row)
```

```python
#  Write a program to multiply two matrices
```

```python
def multiply_matrices(A, B):
    n, m, p = len(A), len(A[0]), len(B[0])
    result = [[0 for _ in range(p)] for _ in range(n)]
    for i in range(n):
        for j in range(p):
            for k in range(m):
                result[i][j] += A[i][k] * B[k][j]
    return result
A = [
    [1, 2, 3],
    [4, 5, 6]
]   

B = [
    [7, 8],
    [9, 10],
    [11, 12]
]   
C = multiply_matrices(A, B)
print("Resultant Matrix (A x B):")
for row in C:
    print(row)
```
